```python
#along with boxplot plotting swarmplot to know no.of outlier points 
col_to_check=['Avg. Session Length','Time on App','Time on Website','Length of Membership']
j=1
plt.figure(figsize=(20,40))
for i in col_to_check:
    plt.subplot(len(col_to_check),2,j)
    sns.boxplot(df[i])
    plt.title(i)
    j=j+1
    plt.subplot(len(col_to_check),2,j)
    sns.swarmplot(df[i])
    j=j+1
```

```python
col_outliers=['Avg. Session Length','Time on App','Time on Website','Length of Membership']
for i in col_outliers:
    q1 , q3 =df[i].quantile([0.25,0.75])
    iqr = q3-q1
    lf , uf = q1-1.5*iqr , q3+1.5*iqr
    df[i]= np.where ( df[i]<lf, df[i].mean(), df[i])
    df[i]= np.where ( df[i]>uf, df[i].mean(), df[i])
```

```python
#along with boxplot plotting swarmplot to know no.of outlier points 
col_to_check=['Avg. Session Length','Time on App','Time on Website','Length of Membership']
j=1
plt.figure(figsize=(20,40))
for i in col_to_check:
    plt.subplot(len(col_to_check),2,j)
    sns.boxplot(df[i])
    plt.title(i)
    j=j+1
    plt.subplot(len(col_to_check),2,j)
    sns.swarmplot(df[i])
    j=j+1
```

now our data is ready for model building

# Model building 

```python
X=df.drop('Yearly Amount Spent',axis=1)
y=df['Yearly Amount Spent']
```

```python
X.dtypes
```

```python
#train test split for boxcox skewed dataframe
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#LinearRegression
lin=LinearRegression()
lin.fit(X_train,y_train)
y_pred_lin = lin.predict(X_test)
```

```python
#Training dataset score
from sklearn.metrics import r2_score
y_pred_train=lin.predict(X_train)
r2_score(y_train,y_pred_train)
```

```python
#Testing dataset score
r2_score(y_test,y_pred_lin)
```

```python
#ajusted r square
score=r2_score(y_test,y_pred_lin)
adj_r2 = 1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
adj_r2
```
