Step 1. Import the necessary libraries

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

Step 2. Import the dataset from this address
Step 3. Assign it to a variable titanic

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv",sep=",")
```

Step 4. Set PassengerId as the index

```python
titanic.set_index("PassengerId",inplace=True)
```

Step 5. Create a pie chart presenting the male/female proportion

```python
plt.pie(titanic.Sex.value_counts(),labels=["male","female"])
```

Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.relplot(y="Fare",x="Age",data=titanic,hue="Sex")
```

Step 7. How many people survived?

```python
titanic.Survived.value_counts().tail(1)
```

Step 8. Create a histogram with the Fare payed

```python
sns.histplot(titanic.Fare)
```
